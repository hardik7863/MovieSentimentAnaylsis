Simple RNN consist of two components:-

1) Embedding layer

2) Simple RNN

In this embedding layer we convert the word into vectors it uses various word embedding methods like Word2Vec

![Screenshot from 2025-01-13 17-15-50.png](<attachment:Screenshot from 2025-01-13 17-15-50.png>)

In Embedding layer,if we use one hot encoding then it will create a sparse matrix which lead to overfitting

![Screenshot from 2025-01-13 17-16-55.png](<attachment:Screenshot from 2025-01-13 17-16-55.png>)

One hot encoding is not the efficient way of doing this hence we use another technique called word embedding

Word2Vec is a kind of Word Embedding technique used in Embedding layer

Word Embedding creates feature representation for each words present in the dataset
